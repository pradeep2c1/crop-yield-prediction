## Rain and Temp Data from IMDLib

```python
!pip install imdlib
import imdlib as imd
```

```python
start_yr = 1997
end_yr = 1997
variable = 'rain' # other options are ('tmin'/ 'tmax')
data = imd.get_data(variable, start_yr, end_yr, fn_format='yearwise')
```

```python
ds = data.get_xarray() # Converting to xarray is also required for SLICING, for a date.
ds
```

```python
d1 = ds.sel(time='1997-04-01')
d1
d2 = ds.sel(time='1997-04-02')
d2
```

```python
avg_d = (d1 + d2)/2 # Average rain of 2 days
# Taking the average of 2 days data, similarly, we'll get the average for whole cultivation period
avg_d
```

## Using the `where` method of xarray

```python
#from shapely.wkt import loads
```

```python
!pip install rasterio
```

```python
import rasterio
lon_values = avg_d.lon.values
lat_values = avg_d.lat.values
lon_spacing = lon_values[1] - lon_values[0]
lat_spacing = lat_values[1] - lat_values[0]
transform = rasterio.transform.from_origin(lon_values[0], lat_values[-1], lon_spacing, lat_spacing)
```

```python
import rasterio.features
mask = rasterio.features.geometry_mask(a_shape, out_shape=(avg_d.lat.size, avg_d.lon.size),
                                       transform=transform, invert=True)
```

```python
masked_xarray = avg_d.where(mask)
masked_xarray
```

```python
import numpy as np

mean_variables = {}
for var_name in masked_xarray.data_vars:
    mean_variables[var_name] = np.nanmean(masked_xarray[var_name])

print(mean_variables)

np.set_printoptions(threshold=np.inf)  # Display the entire array without truncation

# Access and print the 'rain' data variable
print(masked_xarray['rain'].where(masked_xarray['rain'] > 0))
```

```python
masked_xarray['rain']

# Check data distribution
rd = masked_xarray['rain'].where(avg_d['rain'] > 0).values
print("Minimum Rain:", np.nanmin(rd))
print("Maximum Rain:", np.nanmax(rd))
print("Average Rain (excluding NaN values):", np.nanmean(rd))

# Plot histogram of temperature values
plt.hist(rd[~np.isnan(rd)], bins=50)
plt.xlabel('Rain')
plt.ylabel('Frequency')
plt.title('Rain Distribution')
plt.show()
```

```python
# import numpy as np
non_zero_count = np.count_nonzero((masked_xarray['rain'] > 0) & ~np.isnan(masked_xarray['rain']))
print("Number of data entries lying in a district boundary: ", non_zero_count)
```
